```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import train_test_split
import sklearn
import pandas as pd
import numpy as np
```

```python
condicion_feature = pd.read_csv('features/condition_feature.csv')
```

```python
test_kaggle = pd.read_csv('../fiuba-trocafone-tp2-final-set/trocafone_kaggle_test.csv')
```

```python
events = pd.read_csv('../fiuba-trocafone-tp2-final-set/events_up_to_01062018.csv',low_memory=False)
```

```python
condicion_feature
```

```python
knn = KNeighborsClassifier()

X = np.array(condicion_feature[['Bom','Bom - Sem Touch ID','Excelente','Muito Bom','Novo']])
y = np.array(condicion_feature['label'])

X_train,X_test,Y_train,Y_test = train_test_split(X,y)
```

```python
X_test
```

```python
knn.fit(X_train,Y_train)
```

```python
predic = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(predic,Y_test))
```
