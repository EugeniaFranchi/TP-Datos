```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import train_test_split
import sklearn
import pandas as pd
import numpy as np

from time import time
```

```python
condicion_feature = pd.read_csv('features/condition_feature.csv')
```

```python
condicion_general = pd.read_csv('features/condicion_general.csv')
```

```python
promedio_visitas = pd.read_csv('features/promedio_visitas_por_dia.csv')
```

```python
training = pd.read_csv('../fiuba-trocafone-tp2-final-set/labels_training_set.csv',low_memory=False)
```

```python
test_kaggle = pd.read_csv('../fiuba-trocafone-tp2-final-set/trocafone_kaggle_test.csv')
```

```python
events = pd.read_csv('../fiuba-trocafone-tp2-final-set/events_up_to_01062018.csv',low_memory=False)
```

```python
train_df = promedio_visitas.merge(training,on='person',how='right')
```

```python
train_data = pd.read_csv('features/train_data.csv')
```

```python
conversiones_en_6_meses = pd.read_csv('features/conversiones_en_6_meses.csv')
```

```python
train_data = train_df.merge(condicion_general,on='person',how='inner')
```

```python
condicion_general
```

```python
train_data
```

```python
train_data.to_csv('features/train_data.csv')
```

```python
knn = KNeighborsClassifier()

X = np.array(train_data[['Bom','event','mean_conversion']])
y = np.array(train_data['label'])

X_train,X_test,Y_train,Y_test = train_test_split(X,y)
```

```python
knn.fit(X_train,Y_train)
```

```python
predic = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(predic,Y_test))
```
