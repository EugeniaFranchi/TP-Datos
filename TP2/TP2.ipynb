# Feature : Condicion del producto

```python
events_training = events.merge(training,on='person',how='inner')
```

```python
events_training
```

```python
events_training = events_training[['person','event','condition']]
```

```python
events_training = events_training.fillna(value = 0)
```

```python
events_training
```

```python
condicion = events_training.groupby(['person','condition']).count()
condicion
```

```python
condicion_unstack = condicion.unstack()
condicion_unstack
```

```python
condicion = condicion_unstack['event']
```

```python
condicion.reset_index(inplace=True)
condicion.fillna(value=0)
```

```python
condicion_training = condicion.merge(training,on='person',how='left')
condicion_training.fillna(value=0,inplace=True)
```

```python
condicion_training = condicion_training.astype(dtype={'Bom':np.float32,'Bom - Sem Touch ID':np.float32,'Excelente':np.float32,'Muito Bom':np.float32,'Novo':np.float32})
```

```python
condicion_training.to_csv('features/condition_feature.csv')
```
