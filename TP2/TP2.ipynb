# Feature : Condicion del producto

```python
events['condition'].value_counts()
```

```python
condicion = events.loc[events['condition'].isnull() == False]
condicion
```

```python
encoder = LabelBinarizer()
encoder.fit(condicion['condition'])
```

```python
cod = encoder.transform(condicion['condition'])
```

```python
cod
```

```python
condicion.reset_index(inplace = True)
```

```python
encoding = pd.DataFrame(cod)
encoding
```

```python
#Como son 5 tipos de 'condition', el codificador usa 5 bits
condicion['encoding0'] = encoding[0]
condicion['encoding1'] = encoding[1]
condicion['encoding2'] = encoding[2]
condicion['encoding3'] = encoding[3]
condicion['encoding4'] = encoding[4]
```

```python
merge = training.merge(condicion,on='person',how='left')
```

```python
features = merge.loc[merge['encoding0'].isnull() == False]
```

```python
#A medida que agregamos features al tp lo guardamos en un .csv
features
```

```python
clasifier = tree.DecisionTreeClassifier()

X = np.array(features[['encoding0','encoding1','encoding2','encoding3','encoding4']])
y = np.array(features['label'])

X_train,X_test,Y_train,Y_test = train_test_split(X,y)
```

```python
clasifier.fit(X_train,Y_train)
```

```python
predic = clasifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(Y_test,predic))
```
