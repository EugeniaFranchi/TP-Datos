# Feature: Cantidad promedio de compras por usuario por mes (en 6 meses)

```python
conversiones = events[['timestamp','person','event']]
```

```python
conversiones = conversiones.loc[conversiones['event']=='conversion']
```

```python
conversiones['timestamp'] = pd.to_datetime(conversiones['timestamp'])
conversiones['mes'] = conversiones['timestamp'].dt.month
```

```python
conversiones
```

```python
df = conversiones.groupby(['mes','person']).count()
df.reset_index(inplace=True)
conversiones_en_6_meses = df.groupby(['person']).agg({'timestamp':'sum'})
conversiones_en_6_meses
```

```python
conversiones_en_6_meses['mean_conversion'] = conversiones_en_6_meses['timestamp']/6
```

```python
conversiones_en_6_meses
```

```python
train_data = train_data.merge(conversiones_en_6_meses,on='person',how='left')
```

```python
train_data['mean_conversion'].fillna(value=0,inplace=True)
train_data
```

```python
train_data.to_csv('features/train_data.csv')
```

# Feature: Cantidad de eventos por semana

```python
eventos_por_semana = events[['timestamp','person','event']]
```

```python
eventos_por_semana['week'] = pd.to_datetime(eventos_por_semana['timestamp']).dt.week
```

```python
eventos_por_semana = eventos_por_semana.groupby(['person','week','event']).agg({'timestamp':'count'})
```

```python
eventos_por_semana = eventos_por_semana.reset_index()
```

```python
eventos_por_semana = eventos_por_semana.groupby(['person','event']).agg({'timestamp':'mean'}).unstack()
```

```python
eventos_por_semana.reset_index(inplace=True)
```

```python
eventos_por_semana.columns= [col[1] for col in eventos_por_semana.columns]
```

```python
eventos_por_semana.rename(columns={'':'person'},inplace=True)
eventos_por_semana.fillna(value=0,inplace=True)
```

```python
train_data = train_data.merge(eventos_por_semana,on='person',how='left')
```

```python
train_data.fillna(value=0,inplace=True)
train_data.drop(columns=['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1'],inplace=True)
```

```python
test_kaggle = test_kaggle.merge(eventos_por_semana,on='person',how='left')
```
